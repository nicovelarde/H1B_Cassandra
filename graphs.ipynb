```python
import pandas as pd

# Initialize a dictionary to store the number of applications per year
applications_per_year = {}

# Loop through each year from 2011 to 2023
for year in range(2011, 2024):
    # Construct the file name based on the year
    file_name = f'{year}_data.csv'
    
    # Load the data from the file
    data = pd.read_csv(file_name)
    
    # Count the number of rows (assuming the first row is headers)
    applications_per_year[year] = len(data)

# Convert the dictionary to a pandas Series for easy plotting
application_counts = pd.Series(applications_per_year)
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(12, 6))
application_counts.plot(kind='line', marker='o', linestyle='-', color='navy')
plt.title('Number of Applications Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(application_counts.index, application_counts.values, marker='o', linestyle='-', color='navy')

# Annotate each point with the number of applications
for year, count in application_counts.items():
    plt.text(year, count, f' {count}', color='black', ha='center', va='bottom')

plt.title('Number of Applications Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Applications')
plt.grid(True)
plt.xticks(application_counts.index)  # Ensure all years are shown as x-axis ticks
plt.tight_layout()  # Adjust layout to make room for annotations
plt.savefig('applications_per_year.png')
plt.savefig('applications_per_year.jpg')
plt.show()

```
